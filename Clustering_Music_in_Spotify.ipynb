```python
data2 = data[['Beats Per Minute (BPM)', 'Loudness (dB)', 'Liveness', 'Valence', 'Acousticness', 'Speechiness']]
data2.head()
```

# Clustering Analysis of Audio Features

```python
from sklearn.preprocessing import RobustScaler
for column in data2.columns:
    data2[column] = data2[column].astype('float64')
    data2[column] = RobustScaler().fit_transform(data2[[column]])
data2
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 10)
clusters = model.fit_predict(data2)
data['Music Segments'] = clusters
data.head()
```

```python
data['Music Segments'] = data['Music Segments'].map({0: 'Cluster 0',
                                                     1: 'Cluster 1',
 2: 'Cluster 2',
 3: 'Cluster 3',
 4: 'Cluster 4',
 5: 'Cluster 5',
 6: 'Cluster 6',
 7: 'Cluster 7',
 8: 'Cluster 8',
 9: 'Cluster 9',
 10: 'Cluster 10'})
data
```

# Plot Clusters

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

columns = ['Cluster 0', 'Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4', 'Cluster 5', 'Cluster 6', 'Cluster 7', 'Cluster 8', 'Cluster 9']
for i in columns:
    subset = data[data["Music Segments"] == i]
    ax.scatter(subset['Beats Per Minute (BPM)'],
               subset['Energy'],
               subset['Danceability'],
               s=6, label=str(i))

ax.set_xlabel('Beats Per Minute (BPM)')
ax.set_ylabel('Energy')
ax.set_zlabel('Danceability')
ax.set_title('3D Scatter Plot')

plt.legend()
plt.show()
```
